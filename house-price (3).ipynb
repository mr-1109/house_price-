```python
# function for one hot encoding
def cat_onehot_multcols(col):
    df_final=final_data
    i=0
    for fields in col:
        
        print(fields)
        df1=pd.get_dummies(final_data[fields],drop_first=True)
        
        final_data.drop([fields],axis=1,inplace=True)
        if i==0:
            df_final=df1.copy()
        else:
            
            df_final=pd.concat([df_final,df1],axis=1)
        i=i+1
       
        
    df_final=pd.concat([final_data,df_final],axis=1)
        
    return df_final
```

```python
final_data=cat_onehot_multcols(cate_col)
```
